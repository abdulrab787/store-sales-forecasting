```python
import pandas as pd
import joblib
import numpy as np

df = pd.read_csv("../data/processed/train_final.csv")

xgb_model = joblib.load("../models/xgb_model.pkl")
lgb_model = joblib.load("../models/lgb_model.pkl")
stack_model = joblib.load("../models/stacking_meta_model.pkl")
```

```python
df = df.sort_values("date")
split_idx = int(len(df) * 0.8)
val_df = df.iloc[split_idx:]

X_val = val_df.drop(columns=["sales", "date"])
y_val = val_df["sales"]
```

```python
xgb_preds = xgb_model.predict(X_val)
lgb_preds = lgb_model.predict(X_val)

stack_features = pd.DataFrame({
    "xgb_pred": xgb_preds,
    "lgb_pred": lgb_preds
})

stack_preds = stack_model.predict(stack_features)
```

```python
from sklearn.metrics import mean_squared_error

print("XGB RMSE:", np.sqrt(mean_squared_error(y_val, xgb_preds)))
print("LGB RMSE:", np.sqrt(mean_squared_error(y_val, lgb_preds)))
print("STACKING RMSE:", np.sqrt(mean_squared_error(y_val, stack_preds)))
```
