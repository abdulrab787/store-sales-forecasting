```python
import pandas as pd
import joblib
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/train_final.csv")
model = joblib.load("../models/ridge_baseline.pkl")
```

```python
#Time-based split
df = df.sort_values("date")
split_idx = int(len(df) * 0.8)
val_df = df.iloc[split_idx:]

X_val = val_df.drop(columns=["sales", "date"])
y_val = val_df["sales"]

preds = model.predict(X_val)
```

Plot actual vs predicted

```python
plt.figure(figsize=(10,5))
plt.plot(y_val.values[:200], label="Actual")
plt.plot(preds[:200], label="Predicted")
plt.legend()
plt.title("Baseline Predictions vs Actual")
plt.show()
```
